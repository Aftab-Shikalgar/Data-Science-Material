```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
beml = pd.read_csv("Datasets\\BEML.csv")
glaxo = pd.read_csv("Datasets\\GLAXO.csv")
```

```python
beml
```

```python
glaxo
```

```python
beml = beml.set_index(pd.DatetimeIndex(beml['Date']))
glaxo = glaxo.set_index(pd.DatetimeIndex(glaxo['Date']))
```

```python
beml
```

# glaxo

```python
plt.figure(figsize=(12,3))
plt.subplot(1,2,1)
plt.plot(beml.Close)
plt.title("BEML")
plt.subplot(1,2,2)
plt.plot(glaxo.Close)
plt.title("GLAXO")
```

```python
# adding new column, persectage gain

beml['gain'] = beml.Close.pct_change(periods=1)
glaxo['gain'] = glaxo.Close.pct_change(periods=1)
```

```python
beml
```

```python
glaxo
```

```python
beml = beml.dropna()
glaxo = glaxo.dropna()
```

```python
beml
```

```python
glaxo
```

```python
plt.figure(figsize=(12,3))
plt.subplot(1,2,1)
plt.plot(beml.gain)
plt.title("BEML")
plt.subplot(1,2,2)
plt.plot(glaxo.gain)
plt.title("GLAXO")
```

```python
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

plt.figure(figsize=(12,3))
plt.subplot(1,2,1)
sns.distplot(beml.gain)
plt.title("BEML")
plt.subplot(1,2,2)
sns.distplot(glaxo.gain)
plt.title("GLAXO")
```

```python
plt.figure(figsize=(10,5))
sns.distplot(glaxo.gain,label="GLAXO")
sns.distplot(beml.gain,label="BEML")
plt.legend()
```

```python
# mean and sd for BEML
print("BEML mean=",round(beml.gain.mean(),4))
print("BEML standard deviation=",round(beml.gain.std(),4))
```

```python
# mean and sd for GLAXO
print("GLAXO mean=",round(glaxo.gain.mean(),4))
print("GLAXO standard deviation=",round(glaxo.gain.std(),4))
```
