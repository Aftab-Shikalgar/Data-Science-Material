<h2 style="color:red">Divide and Conquer -</h2>

- __Decision trees__ are built using a heuristic called __recursive partitioning__
- This approach is commonly known as __Devide and Conquer__ because it splits data into subsets, which are then split
  repeatedly into even samller subsets and so on until the process stops
- At this point the algorithm determines the data within the subsets are sufficiently homogenous or another stopping criterion
  has been met

<h2 style="color:red">How Decision Tree Works?</h2>

- Imagine a bare root node that will grow into a mature tree
- At first, the root node represents the entire dataset, since no splitting is done
- Next, the decision tree algorithm must choose a feature to split upon
- Ideally it chooses the feature most predictive of the target class.
- The examples then are partitioned into groups according to the distinct values of this feature, and the first set of tree branches is formed

<h1 style="color:darkblue">Decision Tree - C 5.0 -</h1>


<h2>C 5.0 is one of the best implementations of Decision Tree Algorithm</h2>


- Can construct accurate and interpretable decision tree models
- Uses pruning operations to enhance model generalization
- Can be used for classification tasks
- The input features may be continuous or categorical, and the output must be categorical
- The importance of a variable is equal to the proportion of observations that have been classified using that variable

<h2>Tree Building Methodology -</h2>

- The first question is which feature to select first to start with?
- There are various measures to identify best decision tree splitting candidate
- __C 5.0__ uses __Entropy__, a concept borrowed from information theory __that quantifies the randomness or disorder
  within a set of class values__
- __Data sets with high entropy are very diverse and provide little information__ about other items that may also belong
  in the set, as there is no apparent commonality
- Decision tree hopes to find splits that reduce Entropy

<h2>Entropy -</h2>

- Typically, measured in __bits__
- If there are only 2 possible classes, entropy values can range from 0 to 1
- For n classes, entropy ranges from __0 to log2(n)__
- Minimum value indicates that the sample is completely homogenous, while the maximum value indicates that the data are as diverse as possible

<h2>Information Gain -</h2>

- To use entropy to determine the optimal feature to split upon, the algorithm calculates the change in homogeneity that would<br> 
  result from a split on each possible feature, which is a measure known as __Information Gain__

<h2>Pruning The Decision Tree -</h2>

- __A decision tree can continue to grow indefinitely__, choosing splitting features and dividing into smaller and smaller groups <br>
    until each example is perfectly classified  or algo runs out of features to split on
  
- If the tree grows overly large, many of the decisions it makes will be overly specific and the model will be __overfitted__ to the training data

- The process of pruning a decision tree involves reducing its size such that it generalizes better to unseen data

<h1 style="color:darkblue">Classification And Regression Tree(CART) -</h1>


- uses Gini Impurity to split data into a decision tree
- Finds the best homogeneity for the sub nodes
- Can be used for classification and regression tasks
- Minimizes the variance of the target variable inside each subset when deciding which subsets to split

<h2>Gini Impurity for classification problems -</h2>

- __The Gini Impurity of the node__ is the probability that a randomly chosen sample in a node would be incorrectly labelled if it was <br>
  labelled by the distribution of samples in the node
- __e.g.__ In the top(root) node, there is 44.4% chance of incorrectly classifying a data point chosen at random based on the sample labels <br>
  in the node
- We arrive at this value using the following eqn:
  
  <img src="images\\CART_GI_Formula.png" alt='image' width="200px" style="margin-left:30%">
- The Gini Impurity of a node __n__ is 1 minus the sum over all the classes __J__(for binary classification this is 2) of the fraction of <br>
  examples in each class p_i squared

    <img src="images\\CART_GI_Formula1.png" alt='image' width="300px" style="margin-left:30%">
- At each node, the decision tree searches through the features for the value to split on that results in the __greatest reduction in Gini Impurity__

# Example Code

```python
import pandas as pd
import matplotlib.pyplot as plt
#from sklearn import datasets
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import classification_report
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("Datasets\\Iris.csv")
df
```

```python
df.Species.value_counts()
```

```python
#Complete Iris dataset. Since Y variable is categorical convert it to numerical using LabelEncoder
# If Independent variable would have been categorical we would have used one hot encoding.

label_encoder = LabelEncoder()

df.Species = label_encoder.fit_transform(df.Species)

df
```

<h2>Split data into X and Y</h2>

```python
x = df.iloc[:,1:5]
y = df.iloc[:,5] # or df['Species']
```

```python
x
```

```python
y
```

```python
# Column names

colnames = list(df.columns)
colnames
```

## Split entire data in train data and test data

```python
# Splitting data into training and testing data set. 20% data kept aside for testing
# random_state=40: To make splitting consistent across multiple runs.
# output will be 4 dataframes: xtrain, xtest, ytrain and ytest
x_train, x_test,y_train,y_test = train_test_split(x,y, test_size=0.2,random_state=0)
```

```python
x_train
```

```python
y_train
```

```python
y_test
```

```python
y_test.shape
```

## Building Decision Tree Classifier using Entropy Criteria

```python
model = DecisionTreeClassifier(criterion='entropy',random_state=0)    # C 5.0 but without max depth
# To ensure that you get consistent results every time you run your Decision Tree model,
# you can set the random_state=0 parameter for the Decision Tree itself.
model.fit(x_train,y_train)
```

```python
tree.plot_tree(model)
```

```python
# As we can see, at depth 2 the data was significantly homogenous on every leaf node, but it drew whole tree with splitting 

# This is overfitting which is not accurate

# Let's retry with providing max_depth
```

```python
model = DecisionTreeClassifier(criterion='entropy',random_state=0,max_depth=2)
model.fit(x_train,y_train)
tree.plot_tree(model)
```

```python
# Now it's good
```

```python
y_train.value_counts()
```

```python
# even though the tree is proper, it's hard to interpret 
# lets add feature names and class names to make it more readable

fn=x_train.columns # Feature names
cn=['setosa', 'versicolor', 'virginica'] # Class names
#fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (3,3), dpi=300)
tree.plot_tree(model,
               feature_names = fn,
               class_names=cn,
               filled = True);
```

```python
# predicting on test data

preds = model.predict(x_test)
preds # 1st data point belongs to class0, 2nd to 1, 3rd to 2 and so on
```

```python
pd.Series(preds).value_counts()
```

```python
model.feature_importances_ # shows importance of features, SL SW PL PW, PW is most imp feature.
# same is seen in tree. Petal Width is root node
```

```python
pd.crosstab(y_test,preds)
```

```python
# Accuracy

(11+13+5)/(11+13+5+1)
```

```python
# or
np.mean(preds==y_test)
```

## Building Decision Tree Classifier (CART) using Gini Criteria

```python
gini_model = DecisionTreeClassifier(criterion='gini',random_state=0,max_depth=3)   # only param change is value of criterion
gini_model.fit(x_train,y_train)
```

```python
tree.plot_tree(gini_model)
```

```python
pred = gini_model.predict(x_test)
pred
```

```python
pd.crosstab(y_test,pred)
```

```python
# Accuracy

np.mean(y_test==pred)
```
