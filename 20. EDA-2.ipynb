<h1 style="color:red">Correlation Using Predictive Power Score (PPS) -</h1>

__What will be coefficient of correlation for the following associations??__

<img src="images\\EDA2_D3.png" alt='src' width='700px' style="margin-left:100px">

- If the relationship is linear between the params then we can use correlation coefficient to find the strength of assocition
- What if relationship is non-linera, gaussian or unknown relationship like imgs above, how do we find the association?
- What if we want to detect __relationship between cities and zip codes ?__
- The expectation is that, irrespective of the relationship between the params we would like to know the relationship
- And the __score should be 0 if there is no relationship and the score should be 1 if there is a perfect relationship__
- Also score __should be able to handle categoric and numeric columns__

<h2 style="color:limegreen">Predictive Power Score (PPS) -</h2>

- PPS is an asymmetric, data-type-agnostic score that can detect linear or non-linear relationships between 2 columns
- __Score ranges from 0 (no predicive power) to 1 (perfect predictive power)__
- It can be used as an __alternative to the correlation matrix__

- We can now calculate a __cross-validated Decision Tree__ and __calculate suitable evaluation metric__
- When the target is numeric we can use a __Decision Tree Regressor__ and calculate the __Mean Absolute Error(MAE__)
- When the target is catagorical, we can use __Decision Trre Classifier__ and calculate the __weighted F1__

<h3 style="color:blue">Example: Zip Codes and City Names</h3>

- Calculate PPS of zip code to city
- Weighted __F1 score__ will be used because __city is categoric__
- Let's say cross-validated __Decision Tree Classifier__ achieves a score of __0.95 F1__
- Then calculate baseline score via always predicting the most common city and achieve a score of __0.1 F1__
- Normalize the score, you will get final PPS of 0.94 after applying following normalization formula __(0.95-0.1)/(1-0.1)__
- As we can see, a PPS score of 0.94 is rather high, so __zip codes seems to have good predictive power towards city__
- However if we calculate PPS in opposite direction, we might have PPS of close to 0
