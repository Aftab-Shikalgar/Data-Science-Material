## Association Rules

<h4 style="color:orange">Association rule learning is a type of unsupervised learning technique that checks for the dependency of one data item on another data item and maps accordingly</h4>

- Basically it is study of "What goes with what"
- Also called as __market basket analysis,affinity analysis__
- Origin -> Study of customer transaction databases to determine dependencies between purchases of different items

<h3 style="color:red">How Association Rules can be used?</h3>

- Diapers and Beer. Most famous example of market basket analysis for the last few years.
  __If you buy diapers, you tend to buy beer__
- Promotion on one item, raise price of related item
- To determine placement of items in a store
- Stocking -> The process of putting items on shelves and display areas in a store, or in a warehouse or backroom

- __Inexplicable Rules__ - No explanation and do not suggest action

<h3 style="color:red">Rule Generation - </h3>

- __Problem__ - Generating all possible rules is exponential in the number of distinct items

- __Solution__ - __Frequent Item Sets__

              - Consider only combinations that occur with higher frequency in the database
              - Criterion for "frequent" --> Support

<h2 style="color:firebrick">Generating frequent item sets: The Apriori Algorithm</h2>

__For k products -__

1. --> <b>Set minimum support criterion</b>
2. --> <b>Generate list of one-item sets that meet the support criterion</b>
3. --> <b>Use list of one-item sets to generate list of two-item sets that meet support criterion</b>
4. --> <b>Use list of two-item sets to generate list of three-item sets that meet support criterion</b>
5. --> <b>continue up through k-item sets</b>

<h3 style="color:red">Performance Measure - Lift Ratio</h3>

- __Lift_ratio = confidence / (benchmark confidence)__

- benchmark assumes independence between antecedent and consequence

- <img src="images\\Association_Rules_Lift.png" alt="image" width="400px">

<h3 style="color:red">Interpreting Lift -</h3>

<h3>Lift > 1 indicates a rule that is useful in finding consequent item sets i.e. more useful than selecting transactions randomly</h3>

# Apriori Algorithm

```python
frequent_itemsets = apriori(df, min_support=0.1, use_colnames=True)#df-dataframe, min_support default=0.5
# min_support is a float between 0 and 1 for minumum support of the itemsets returned.
# The support is computed as the fraction transactions_where_item(s)_occur / total_transactions.
frequent_itemsets # output is frequent itemsets: 1-item itemsets, 2-item itemsets etc. with min support criteria
```

```python
rules = association_rules(frequent_itemsets,metric='lift',num_itemsets=2201,min_threshold=0.7)   # min_threshold is confidence
rules.sort_values('lift',ascending=False)[0:20]    # first 20
```

```python
rules[rules.lift>1]
```

# Another Example with Retail_Dataset

```python
retail = pd.read_csv("Datasets\\retail_dataset.csv")
retail
```

```python
retail.shape
```

```python
retail['0'].unique()
```

```python
items = retail['1'].unique()
```

```python
for i in retail.iterrows():
    print(i)
```

```python
# One-hot encoding
encoded_vals = []
for index, row in retail.iterrows(): # iterrows() returns index and content for each row
    labels = {}
    uncommons = list(set(items) - set(row))#total 9 items (Bagel,Milk,...,pencil) - items present in row = Bagel, Milk (for 1st row these 2 items are not present)
    commons = list(set(items).intersection(row))# common items
    for uc in uncommons:
        labels[uc] = 0 # assign 0 to uncommon items
    for com in commons:
        labels[com] = 1# assign 1 to common items
    encoded_vals.append(labels)
```

```python
encoded_vals
```

```python
ohe_df = pd.DataFrame(encoded_vals)
```

```python
ohe_df
```

```python
ohe_df.drop(columns=ohe_df.columns[0],inplace=True)
ohe_df
```

```python
freq_items = apriori(ohe_df,min_support=0.2,use_colnames=True)
```

```python
freq_items
```

```python
asso_rules = association_rules(freq_items,metric='confidence',min_threshold=0.6,num_itemsets=315)  # can use metric as lift
asso_rules.sort_values('confidence',ascending=False)
```

```python
asso_rules[asso_rules.confidence>0.6]
```
